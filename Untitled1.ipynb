```python
import requests
import json
from shapely.geometry import LineString
from shapely.ops import transform
import pyproj
from functools import partial

# Define USGS API URLs
api_key_url = "https://earthexplorer.usgs.gov/inventory/json/v/1.4.0/login"
search_url = "https://earthexplorer.usgs.gov/inventory/json/v/1.4.0/search"
download_url = "https://earthexplorer.usgs.gov/inventory/json/v/1.4.0/download"
logout_url = "https://earthexplorer.usgs.gov/inventory/json/v/1.4.0/logout"

# Replace with your USGS credentials
username = "AdarshaReddyAdi"
password = "adireddy@123456"

# Define the coordinates for the line between your sites (e.g., river stretch)
sites = [
    (76.879, 30.450),  # Site 1
    (77.277, 28.613),  # Site 2
    (77.503, 27.215),  # Site 3
    (78.008, 27.176),  # Site 4
    (81.843, 25.435)   # Site 5
]

# Create a buffer zone around the river line (10 km buffer in meters)
line = LineString(sites)
buffer_distance = 10000  # 10 km in meters

# Define the coordinate transformations properly
wgs84 = pyproj.CRS('EPSG:4326')
# UTM zone 43N (since your coordinates are in northern India)
utm = pyproj.CRS('EPSG:32643')

# Create the projections
project = pyproj.Transformer.from_crs(wgs84, utm, always_xy=True).transform
project_back = pyproj.Transformer.from_crs(utm, wgs84, always_xy=True).transform

# Apply the transformations
line_utm = transform(project, line)
buffer_polygon_utm = line_utm.buffer(buffer_distance)
buffer_polygon = transform(project_back, buffer_polygon_utm)

# Convert polygon to bounding box (min lat/lon and max lat/lon)
min_lon, min_lat, max_lon, max_lat = buffer_polygon.bounds

# Login and get API key
response = requests.post(api_key_url, json={"username": username, "password": password})
if response.status_code == 200:
    try:
        api_key = response.json().get('data')
        if not api_key:
            print("Login failed: No API key returned.")
            exit()
    except json.JSONDecodeError:
        print("Error decoding JSON response from login request.")
        exit()
else:
    print(f"Login failed with status code {response.status_code}. Response text: {response.text}")
    exit()

# Define search parameters for Landsat 5 TM and Landsat 8 OLI
datasets = {
    "LANDSAT_TM_C1": ["4", "3", "2"],  # Band 4, 3, 2 for FCC
    "LANDSAT_8_C1": ["7", "6", "4"]    # Band 7, 6, 4 for FCC
}

# Define temporal range and spatial filters
temporal_filter = {
    "startDate": "2019-01-01",
    "endDate": "2023-01-01"
}

# Loop through datasets and search for images
for dataset, bands in datasets.items():
    search_params = {
        "apiKey": api_key,
        "datasetName": dataset,
        "spatialFilter": {
            "filterType": "mbr",
            "lowerLeft": {"latitude": min_lat, "longitude": min_lon},
            "upperRight": {"latitude": max_lat, "longitude": max_lon}
        },
        "temporalFilter": temporal_filter,
        "maxResults": 10,
        "startingNumber": 1,
        "sortOrder": "DESC"
    }

    search_response = requests.post(search_url, json=search_params)
    if search_response.status_code == 200:
        try:
            search_results = search_response.json().get("data", {}).get("results", [])
            # Loop through each search result and download specified bands
            for item in search_results:
                entity_id = item["entityId"]
                print(f"Found {dataset} image with ID: {entity_id}")

                # Request download links for specific bands
                download_params = {
                    "apiKey": api_key,
                    "datasetName": dataset,
                    "entityId": entity_id,
                    "products": ["STANDARD"]  # "STANDARD" product includes all bands
                }
                
                download_response = requests.post(download_url, json=download_params)
                if download_response.status_code == 200:
                    try:
                        download_data = download_response.json().get("data", {})
                        if download_data:
                            # Extract download links for each band
                            for band in bands:
                                band_link = download_data.get("downloadUrls", {}).get(f"band{band}")
                                if band_link:
                                    print(f"Downloading Band {band} for {entity_id}: {band_link}")
                                    # Uncomment to download the band directly
                                    # band_data = requests.get(band_link)
                                    # with open(f"{entity_id}_B{band}.tif", "wb") as f:
                                    #     f.write(band_data.content)
                    except json.JSONDecodeError:
                        print("Error decoding JSON response for download request.")
                else:
                    print(f"Failed to get download link with status code {download_response.status_code}")
        except json.JSONDecodeError:
            print("Error decoding JSON response for search request.")
    else:
        print(f"Search request failed with status code {search_response.status_code}")

# Logout from the API
try:
    requests.post(logout_url, json={"apiKey": api_key})
    print("Logged out successfully.")
except Exception as e:
    print(f"Error during logout: {e}")
```
